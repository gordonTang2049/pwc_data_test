Purpose of this road accident analysis is to understand the data of road accidents and assist of allocating resource to reduce road accidents.
 

## Top 10 Highest Number of Road Accidents by Region

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
acc_infor = pd.read_csv('ACCIDENT.csv')
acc_loc = pd.read_csv('ACCIDENT_LOCATION.csv')
node = pd.read_csv('NODE.csv')
mer_acc_infor_node = acc_infor.merge(node, on='ACCIDENT_NO', how='left')
series = mer_acc_infor_node[['LGA_NAME']]\
    .groupby(['LGA_NAME']).LGA_NAME.count().sort_values(ascending=False)

seriesh = series.head(10)
areas = seriesh.keys()

sns.set_style('darkgrid')
x = areas
y = seriesh.values
plt.figure(figsize=(16, 6))
sns.barplot(x, y)\
    .set_title("Top 10 of highest Road Accidents Region")
plt.xlabel("Name of Region", size=10)
plt.ylabel("Number of Accident", size=10)
print(seriesh)
plt.show()
```

<p>The bar chart shows that Melbourne has the highest number of road accidents,</p> followed by Casey, which is in greater Melbourne area. <p>The authority, therefore, should invest most of their resource to reduce chances of accidents. </p> 

## Least 10 of Road Accidents by Region

```python
seriest = series.tail(10)
areas = seriest.keys()
sns.set_style('darkgrid')
x = areas
y = seriest.values
plt.figure(figsize=(16, 6))
sns.barplot(x, y)\
    .set_title("Bottom 10 of lowest Road Accidents Region")
plt.xlabel("Name of Region", size=10)
plt.ylabel("Number of Accident", size=10)
print(seriest)
plt.show()
```

<p>In contrast, Regional areas have significant less accidents than the metropolitan areas. The government ma consider invest less resources to those areas.</p>

## Top 10 Incidents by code

```python
series_dca = acc_infor[['DCA_CODE']]\
    .groupby(['DCA_CODE']).DCA_CODE.count().sort_values(ascending=False)
top10_series_dca = series_dca.head(10)
dca_code = top10_series_dca.keys()
x = dca_code
y = top10_series_dca.values
plt.figure(figsize=(16, 6))
sns.barplot(x, y)\
    .set_title("Top 10 incidents")
plt.xlabel("Name of incidents by DCA code", size=10)
plt.ylabel("Number of Accident", size=10)
print(top10_series_dca)
plt.show()
```

## Accidents Distribution on Weekday by code

```python
number_of_acci_by_weekday = acc_infor\
    .groupby(['DAY_OF_WEEK']).DAY_OF_WEEK.count().sort_values(ascending=False)
weekday_code = number_of_acci_by_weekday.keys()
x = weekday_code
y = number_of_acci_by_weekday.values
plt.figure(figsize=(16, 6))
sns.barplot(x, y)\
    .set_title("Number of Accidents on each weekday")
plt.xlabel("Code of the weekday", size=10)
plt.ylabel("Number of Accidents", size=10)
print(number_of_acci_by_weekday)
plt.show()
```


| Weekday code       | Weekday Description     |
| :------------- | :----------: | 
| 0 | Sunday |
| 1 | Monday |
| 2 | Tuesday |
| 3 | Wednesday |
| 4 | Thursday |
| 5 | Friday |
| 6 | Saturday |

<p> Except Sunday, accidents are fairly distributed on all other days. Department may reduce their resource spending on Sunday. </p> 

## Number of Accidents by speed zone

```python
number_of_accidents_on_speed_zone = acc_infor\
    .groupby(['SPEED_ZONE']).SPEED_ZONE.count().sort_values(ascending=False)
speed_zone_code = number_of_accidents_on_speed_zone.keys()
x = speed_zone_code
y = number_of_accidents_on_speed_zone.values
plt.figure(figsize=(16, 6))
sns.barplot(x, y)\
    .set_title("Number of Accidents on each Speed Zone Code")
plt.xlabel("Code of the Speed Zone", size=10)
plt.ylabel("Number of Accidents", size=10)
print(number_of_accidents_on_speed_zone)
plt.show()
```

<p>Speed limited of 60 has the largest amount of accidents occurred, followed by Speed limited of 60,</p> <p>compared to speed limited of 80s and 100.
